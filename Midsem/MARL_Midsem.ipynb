```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("rewards.csv")
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Episode'], df['Episodic Reward'], label='Episodic Reward', color='r', marker='o', linestyle='--', alpha=0.7)
plt.plot(df['Episode'], df['Cumulative Reward'], label='Cumulative Reward', color='b', marker='*')
plt.title('Episode vs Cumulative Reward')
plt.ylabel('Reward')
plt.xlabel('Episode')
plt.grid()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Number of rows and columns for the grid
rows, cols = 2, 2

# Total points per subplot (250 points per grid)
points_per_subplot = 250

# Create subplots with 2x2 grid
fig, axes = plt.subplots(rows, cols, figsize=(24, 20))  # Adjust the figsize for a larger plot

# Flatten axes for easy iteration
axes = axes.flatten()

# Iterate over each subplot and plot the respective 250 points
for i in range(rows * cols):
    start_idx = i * points_per_subplot
    end_idx = start_idx + points_per_subplot

    # Plot Episodic Reward and Cumulative Reward for the current 250 points
    axes[i].plot(df['Episode'][start_idx:end_idx], df['Episodic Reward'][start_idx:end_idx],
                 label='Episodic Reward', color='r', linestyle='--', alpha=0.7, marker='o')
    axes[i].plot(df['Episode'][start_idx:end_idx], df['Cumulative Reward'][start_idx:end_idx],
                 label='Cumulative Reward', color='b', marker='*')

    # Set title and labels for each subplot with specified font sizes
    axes[i].set_title(f'Episode {start_idx} to {end_idx}', fontsize=24)
    axes[i].set_xlabel('Episode', fontsize=20)
    axes[i].set_ylabel('Reward', fontsize=20)

    # Increase tick label font size
    axes[i].tick_params(axis='both', which='major', labelsize=20)

    axes[i].grid(True)

    # Show legend only in the first subplot for clarity, with larger font size
    if i == 0:
        axes[i].legend(fontsize=14)

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```
